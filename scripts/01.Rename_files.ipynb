# 1. Renaming files

```r
library(dplyr)
library(stringr)
library(tidyr)
```

```r
# Get original file names
rawfile_dir <- "~/Documents/Collaborations/sinusite_edwin/data/raw_data" # specify the directory where the fastq are
rawfiles <- list.files(rawfile_dir, pattern = "\\.fastq$", full.names = TRUE) # list all fastq files
rawfiles
length(unique(rawfiles))

# Define the replacements for time points
time_point_replacements <- c("pré_cirurgia" = "T0", "PreC" = "T0", "1_mês" = "T1", "1M" = "T1", 
                             "3_meses" = "T3", "3M" = "T3", "6_meses" = "T6", "6M" = "T6", 
                             "12_meses" = "T12", "12M" = "T12")

# Create renamed filenames
renamed_filenames <- rawfiles

for (pattern in names(time_point_replacements)) {
  renamed_filenames <- gsub(pattern, time_point_replacements[pattern], renamed_filenames)
}

# Rename files in the directory
file.rename(rawfiles, renamed_filenames)

# Print a message confirming the changes
print("Files have been renamed successfully!")
print(renamed_filenames)
```
